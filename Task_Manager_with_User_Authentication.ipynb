```python
# Shivam Suman Project 2. Task Manager with User Authentication
import hashlib
import os
```

```python
def reg():

  user_name = input("Enter a user_name : ")
# Ensure the user_details file exists else create the file creates an empty file if it doesn't exist.

  if not os.path.exists("user_details.txt"):
    open("user_details.txt", "w").close()

# Check if username already exists to avoid duplicacy
  with open("user_details.txt", 'r') as f:
    for line in f:
      stored_user_name, _ = line.strip().split(",")
      if stored_user_name == user_name:
        print("Username already exists. Try a different one.")
        return reg()  # Retry registration

  password = input("Enter a password : ")

  hash_password = hashlib.sha256(password.encode()).hexdigest() # For hashing the password

  with open("user_details.txt",'a') as f:
    f.write(user_name + "," + hash_password + "\n")

  print("User Registered Sucessfully")

  return user_name

# Login Function

def login():
  user_name = input("Enter your user_name : ")
  password = input("Enter your password : ")

  hash_password = hashlib.sha256(password.encode()).hexdigest()

  with open("user_details.txt",'r') as f:
    for line in f:
      stored_user_name, stored_password = line.strip().split(",")
      if stored_user_name == user_name and stored_password == hash_password:
        print("Login Sucessfully")
        return user_name
  print("Invalid password, Please try again")
  return False

# Add a Task

def add_task(user_name):
    des = input("Enter task description: ")
    file_path = f"{user_name}_tasks.txt"

    # Read existing lines to determine next ID
    try:
        with open(file_path, 'r') as file:
            lines = file.readlines()
            ID = len(lines) + 1
    except FileNotFoundError:
        ID = 1  # Start from 1 if the file doesn't exist

    # Append new task
    with open(file_path, 'a') as file:
        file.write(f"{ID},{des},Pending\n")

    print(f"Task added successfully with ID: {ID}")

# View all the task of the logged in user

def view_task(user_name):
  print(f"\nYour Tasks: ")
  file_path = f"{user_name}_tasks.txt"
  with open(file_path, 'r') as file:
    for line in file:
      task_id, task , status = line.strip().split(',')
      print(f"ID: {task_id}, Task: {task}, Status: {status}")

# Update the task status to Completed

def mark_task(user_name):
  task_id = input("Enter task ID to mark as completed: ")
  file_path = f"{user_name}_tasks.txt"

  tasks = []
  with open(file_path, 'r') as file:
    for line in file:
      task_id_in_file , task , status = line.strip().split(',')
      if(task_id_in_file == task_id):
        status = 'Completed'
      tasks.append(f"{task_id_in_file},{task},{status}")
  with open(file_path, 'w') as file:
    for t in tasks:
      file.write(t + '\n')
  print(f"Task with task id {task_id} marked as completed!")

# Delete any task with the given task ID

def delete_task(user_name):
  task_id = input("Enter task ID to delete: ")
  file_path = f"{user_name}_tasks.txt"
  tasks = []
  with open(file_path, 'r') as file:
    for line in file:
      task_id_in_file , task , status = line.strip().split(',')
      if(task_id_in_file != task_id):
        tasks.append(f"{task_id_in_file},{task},{status}")
  with open(file_path, 'w') as file:
    for t in tasks:
      file.write(t + '\n')
  print(f"Task with task id {task_id} is deleted sucessfully!")

# Create an Interactive Menu:
# â€¢ Build a menu that allows users to choose between:
# o Add a Task
# o View Tasks
# o Mark a Task as Completed
# o Delete a Task
# o Logout


def task_menu(username):

  while True:
    print("\n1. Add Task\n2. View Tasks\n3. Update Task\n4. Delete Task\n5. Logout")
    choice = input("Choose an option: ")
    if choice == "1":
        add_task(username)
    elif choice == "2":
        view_task(username)
    elif choice == "3":
        mark_task(username)
    elif choice == "4":
        delete_task(username)
    elif choice == "5":
        print("Logged out sucessfully!")
        break
    else:
        print("Invalid option.")


def main():
  while True:
    print("\n1. Register\n2. Login Existing User\n3. Exit")
    choice = input("Choose an option: ")
    if choice == "1":
      user = reg()
      if user :
        task_menu(user)
    elif choice == "2":
      user = login()
      if user:
        task_menu(user)
    elif choice == "3":
      print("Thank you for using this application")
      break
    else :
      print("Invalid option.")

if __name__ == "__main__":
  main()







```
