```python
# Shivam Suman Project 1. Personal Expense Tracker
from datetime import datetime
import pandas as pd
import csv
```

```python
# Taking input  from the user for expense details
def add_expense_details(expenses):

  expense_date = input("Enter the date in (YYYY-MM-DD) format: ")
  try:

    date_object = expense_date
  except ValueError:
    print("Invalid date format. Please Enter date in (YYYY-MM-DD) format.")
    return

  category = input("Enter the Category of Purchase (e.g., Food,Travel,Movie,Groceries): ")
  amount = float(input("Enter the amount spent on purchase : "))
  description = input("Enter the description of the purchase :")

  expense = {
      'date': date_object,
      'category': category,
      'amount': amount,
      'description': description
  }
  expenses.append(expense)
  print("Expense added successfully!")

# View expenses function
def view_expenses(expenses):
  if not expenses:
    print("No expenses found.")
  else:
    print("\nExpenses:")
    for expense in expenses:
      if all(key in expense and expense[key] for key in ['date', 'category', 'amount', 'description']):
        print(f"date: {expense['date']}, category: {expense['category']}, amount: {expense['amount']}, description: {expense['description']}")
      else  :
        print("Incomplete expense entry, skipping the row")



# Input and track the budget for the month
def set_budget():
  budget = float(input("Enter your monthy budget :"))
  return budget

def track_expenses(expenses,budget):
  sum = 0
  for expense in expenses:
    sum += expense['amount']

  if (sum > budget):
    print("You have exceeded your budget!")
  else:
    remaining = budget - sum
    print(f"You have {remaining} left for the month)")

# Save the expenses
def save_expenses(expenses,filename = "expenses_raw_data.csv"):
  with open(filename, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['date', 'category', 'amount', 'description'])

    for expense in expenses:
      writer.writerow([expense['date'], expense['category'], expense['amount'], expense['description']])

    print(f"Expenses saved to {filename} sucessfully")

# Read the Expenses and handling the error if file not found
def read_expenses(filename = "expenses_raw_data.csv"):
  expenses = []
  try:
    with open(filename, 'r') as file:
      reader = csv.DictReader(file)
      for row in reader:
        if all(key in row and row[key] for key in ['date', 'category', 'amount', 'description']):
          row['amount'] = float(row['amount'])
          expense = {
              'date': row['date'],
              'category': row['category'],
              'amount': float(row['amount']),
              'description': row['description']
          }
          expenses.append(expense)
        else:
          print(f"Incomplete expense entry, skipping the row {row}")
    print("Expenses loaded from the file sucessfully!")

  except FileNotFoundError:
    print("File not found. No expenses to load.")

  return expenses

def exit():
  print("Thank you for using this application")



#Create an interactive menu:
'''
o Add expense
o View expenses
o Track budget
o Save expenses
o Exit
'''

def main():
  expenses = read_expenses()
  budget = set_budget()
  while True:
    print("\nMenu:")
    print("1. Add expense")
    print("2. View expenses")
    print("3. Track budget")
    print("4. Save expenses")
    print("5. Exit")

    selected_opt = (input("Enter your choice (1-5): "))

    if selected_opt == '1':
      add_expense_details(expenses)
    elif selected_opt == '2':
      view_expenses(expenses)
    elif selected_opt == '3':
      track_expenses(expenses,budget)
    elif selected_opt == '4':
      save_expenses(expenses)
    elif selected_opt == '5':
      save_expenses(expenses) # Always saves the input before Exit.
      exit()
      break
    else:
      print("Invalid choice, please try again")

if __name__ == "__main__":
  main()
```
